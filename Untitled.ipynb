# Initialize the data points into desired number of clusters randomly.

```python
def initializeMembershipMatrix(): # initializing the membership matrix
    membership_mat = []
    for i in range(n):
        random_num_list = [random.random() for i in range(k)]
        summation = sum(random_num_list)
        temp_list = [x/summation for x in random_num_list]

        membership_mat.append(temp_list)
    return membership_mat
```

# Find out the centroid.

 m is the fuzziness parameter (generally taken as 2), and xk is the data point

```python
def calculateClusterCenter(membership_mat): # calculating the cluster center
    cluster_mem_val = list(zip(*membership_mat))   
    cluster_centers = []
    for j in range(k):
        uk = list(cluster_mem_val[j])#get all value for one cluster
        ukm = [y ** m for y in uk]
        denominator = sum(ukm)#sum(yk^m)
        temp_num = []
        for i in range(n):
            data_point = list(df.iloc[i])#zk xk its 4 value
            prod = [ukm[i] * val for val in data_point]#yk^m * xk
            temp_num.append(prod)
        numerator = map(sum, list(zip(*temp_num)))# 4 value"
        center = [z/denominator for z in numerator]
        cluster_centers.append(center)
    return cluster_centers
```
